```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Describing_Distributions.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['VitD_levels'], kde=True, color='blue', label='VitD_levels')
plt.axvline(df['VitD_levels'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['VitD_levels'].median(), color='green', linestyle='--', label='Median')
plt.axvline(df['VitD_levels'].mean() + df['VitD_levels'].std(), color='black', linestyle='--', label='+1 std')
plt.axvline(df['VitD_levels'].mean() - df['VitD_levels'].std(), color='black', linestyle='--', label='-1 std')
plt.axvspan(df['VitD_levels'].mean() - df['VitD_levels'].std(), df['VitD_levels'].mean() + df['VitD_levels'].std(),
            alpha=0.2, color='yellow', label='±1 std')
plt.xlabel('VitD_levels')
plt.legend()
plt.title('Distribution of VitD_levels')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Doc_visits'], kde=True, color='blue', label='Doc_visits')
plt.axvline(df['Doc_visits'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['Doc_visits'].median(), color='green', linestyle='--', label='Median')
plt.axvline(df['Doc_visits'].mean() + df['Doc_visits'].std(), color='black', linestyle='--', label='+1 std')
plt.axvline(df['Doc_visits'].mean() - df['Doc_visits'].std(), color='black', linestyle='--', label='-1 std')
plt.axvspan(df['Doc_visits'].mean() - df['Doc_visits'].std(), df['Doc_visits'].mean() + df['Doc_visits'].std(),
            alpha=0.2, color='yellow', label='±1 std')
plt.xlabel('Doc_visits')
plt.legend()
plt.title('Distribution of Doc_visits')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalCharge'], kde=True, color='blue', label='TotalCharge')
plt.axvline(df['TotalCharge'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['TotalCharge'].median(), color='green', linestyle='--', label='Median')
plt.axvline(df['TotalCharge'].mean() + df['TotalCharge'].std(), color='black', linestyle='--', label='+1 std')
plt.axvline(df['TotalCharge'].mean() - df['TotalCharge'].std(), color='black', linestyle='--', label='-1 std')
plt.axvspan(df['TotalCharge'].mean() - df['TotalCharge'].std(), df['TotalCharge'].mean() + df['TotalCharge'].std(),
            alpha=0.2, color='yellow', label='±1 std')
plt.xlabel('TotalCharge')
plt.legend()
plt.title('Distribution of TotalCharge')
plt.show()
```

```python
kurtosis_value = df['VitD_levels'].kurtosis()

print(f"Kurtosis for 'VitD_levels': {kurtosis_value}")
```
